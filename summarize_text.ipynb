```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
```

```python
openai = OpenAI()

system_prompt = "You are a professional, no nonsense and to the point assitant"
user_prompt = """
    Please generate an email subject line for this email. 
    I hope this email finds you well.
    This is a comprehensive update regarding our project, including several action items and key details.
        1. Project Status Updates
            Module A: Completed on time, awaiting stakeholder review.
            Module B: 50 \%\ complete, delay expected by 2 days due to API issues.
            Budget: Currently \(15\%\) under projected costs.
        2. Key Deliverables Required
            @John - Finalize design assets by Tuesday.
            @David - Update the API documentation by Wednesday.
            @Lisa - Review the Q3 budget spreadsheet.
        3. Action Items & Next Steps
            Meeting: Review session scheduled for Thursday at 10 AM.
            Feedback: Please submit feedback on the draft proposal by Friday EOD.
            Link: The shared document can be accessed here.
        4. Notes & Reminders
        Please ensure all team members use the updated formatting for reports.
        Remember to use the +1 Gmail trick if you need to filter these alerts.
"""

# Step 2: Make the messages list

messages = [{"role": "system" , "content": system_prompt },
            {"role": "user" , "content": user_prompt}] 

# Step 3: Call OpenAI
response = openai.chat.completions.create(
    model = "gpt-4.1-mini" , 
    messages=messages
)

# Step 4: print the result
print(response.choices[0].message.content)
```
