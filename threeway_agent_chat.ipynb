```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
MODEL_LLAMA = 'llama3.2'
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama_a = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
ollama_b = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
ollama_c = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
system_prompt_alex =  """
You are Alex, a chatbot who is very argumentative; you disagree with anything in the conversation and you challenge everything, in a snarky way.
You are in a conversation with Blake and Charlie. Just give your response without any explanations or disclaimers. Keep it to one or two sentences.
"""

def user_promt_alex (conversation) :
    user_prompt = f"""
    You are Alex, in conversation with Blake and Charlie.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as Alex.
    """
    return user_prompt


system_prompt_blake =  """
You are Blake, a chatbot who is very polite, courteous. You try to agree with
everything the other person says, or find common ground. If the other person is argumentative,
you try to calm them down and keep chatting.You are in a conversation with Alex and Charlie. Just give your response without any explanations or disclaimers. Keep it to one or two sentences.
"""

def user_promt_blake(conversation) :
    user_prompt = f"""
    You are Blake, in conversation with Alex and Charlie.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as Blake.
    """
    return user_prompt


system_prompt_charlie =  """
You are Charlie, a chatbot who is neither polite or argumentative. You try to solve conflicts by providing facts, and support the right. 
If the other person is argumentative,
you try to calm them down but still stay on point.
If the orther person is polite, you try to find common ground but still stay on point. You are in a conversation with Alex and Blake. Just give your response without any explanations or disclaimers. Keep it to one or two sentences.
"""

def user_promt_charlie(conversation) :
    user_prompt = f"""
    You are Charlie, in conversation with Alex and Blake.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as Charlie.
    """
    return user_prompt


conversation = ["Charlie: Lets discuss about whether AI is good or bad for humanity?"]

def generate_response(conversation , model):
    if model == 'alex' :
        system_prompt = system_prompt_alex
        user_prompt = user_promt_alex(conversation)
        model_name = ollama_a
    elif model == 'blake' :
        system_prompt = system_prompt_blake
        user_prompt = user_promt_blake(conversation)
        model_name = ollama_b
    else :
        system_prompt = system_prompt_charlie
        user_prompt = user_promt_charlie(conversation)
        model_name = ollama_c  
    
    response = model_name.chat.completions.create(
        model = MODEL_LLAMA ,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    
    return (response.choices[0].message.content)

print (conversation)

for i in range(5):

    for model in ['alex', 'blake', 'charlie']:
        response = generate_response(conversation , model)
        conversation.append(f"{model}: {response}")
        print(f"{model}: {response}\n")
```
